```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_csv.csv')
df
```

```python
# Getting the entries under a particular column in the data frame
df['Actor']
```

```python
df['Actor'][3]
```

```python
type(df['Actor'])
```

```python
# Getting multiple columns
# getting columns like this just returns a view, this does not returns the original dataframe
df[['Actor','IMDb']]
```

```python
# Selecting row by index
df.loc[1]
```

```python
df
```

```python
# Sometimes we change row labels, iloc find on the basis of index
# Loc finds on the basis of the labels
df.iloc[1]
```

```python
# Getting a particular cell
df.loc[11,'Genre']
```

```python
df.iloc[11,3]
```

```python
# Getting multiple rows and in that also getting specific columns
# In loc, here the last index 2 is inclusive, but this is not the case in iloc
df.loc[0:2,['Actor','IMDb','Genre']]
```

```python
# Getting multiple rows and in that also getting specific columns using iloc
# Here the last index is not inclusive
df.iloc[0:2,0:2]
```

```python
# Fast way of accessing element at a particular index (can access only one cell)
# at, iat are optimized to get a single element
df.at[0,'Actor']
```

```python
df.iat[0,0]
```

```python
# Functions for filtering of data in dataframe
# Getting data having IMDb score greater than 6
# df[df['IMDb']>6]['Actor']
df[df['IMDb']>6]
```

```python
# Getting only those films for which IMDb > 6 and Year > 2019
# We can use or condition also (|)
df[(df['IMDb']>6) & (df['Year']>2019)]
```

```python
# df.query uses a natural like language
df.query('Year > 2019')
```

```python
df.query('Year > 2019 and IMDb > 6')
```

```python
# df.query returns a copy instead of a view but for otehr filtering methods Pandas does not guarantees 
# that always a copy will be returned, it can return a view also. If we want a copy, use a .copy method always instead
df2 = df.query('Year > 2019 and IMDb > 6')
df2
```

```python
df2.loc[5,'Actor'] = 'Dhwani Vashisth'
```

```python
df2
```
