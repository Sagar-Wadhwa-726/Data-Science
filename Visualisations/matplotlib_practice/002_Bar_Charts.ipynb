```python
# Bar charts are used to compare quantities across categories
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [1990, 1992,1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
# If we have same data for more than one players

sehwag_runs = [0, 420, 1100, 1250, 1350, 1420, 1100, 1300, 920, 850]
kohli_runs =  [0, 0, 700, 1380, 1650, 1550, 1500, 2595, 2818, 2450]
sachin_runs = [1320, 1894, 1550, 1630, 1410, 1200, 1100, 900, 780, 500]
x = np.arange(len(years))

width = 0.25

plt.bar(x-width, sachin_runs, width=width, label="Sachin")
plt.bar(x, sehwag_runs, width=width, label="Virendar Sehwag")
plt.bar(x+width, kohli_runs, width=width, label="Virat Kohli")
# To show years on the x-axis instead of the indices
plt.xticks(x, years)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
```

```python
# Horizontal Bar Charts
players = ['Sachin', 'Sehwag', 'Kohli', 'Yuvraj']
run_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
plt.barh(players, run_5yrs, color = "skyblue")
plt.xlabel('Total Runs in first 5 years')
plt.ylabel('First 5-Year Performance of Indian Batsmen')
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500,1200,1800]
plt.bar(players, runs, color = 'skyblue')

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i]+50, str(runs[i]), ha='center')

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
years = [2008, 2009, 2010, 2011, 2012]

sachin_scores = [109, 175, 200, 111, 114]
kohli_scores = [85, 107, 118, 183, 133]
sehwag_scores = [119, 131, 173, 73, 96]

x = np.arange(len(years))
width = 0.25

plt.bar(x-width,sachin_scores, width=width, label="Sachin Tendulkar", color='black')
plt.bar(x,kohli_scores,width=width, label="Virat Kohli")
plt.bar(x+width,sehwag_scores,width=width, label="Virendar Sehwag")

for i in range(len(years)):
    plt.text(x[i]-width, sachin_scores[i]+1, str(sachin_scores[i]), ha='center', fontsize=9)
    plt.text(x[i], kohli_scores[i]+1, str(kohli_scores[i]), ha='center', fontsize=9)
    plt.text(x[i]+width, sehwag_scores[i]+1, str(sehwag_scores[i]), ha='center', fontsize=9)


plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.style.use('ggplot')
plt.title("Runs scored by batsmen")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.plot()
```

```python
players = ["Sachin", "Virat", "Virendar"]
run_5yrs = [709, 626, 592]

y = np.arange(len(players))

plt.figure(figsize=(8, 5))
plt.style.use('ggplot')

# Create horizontal bars
plt.barh(y, run_5yrs, color=['black', 'blue', 'green'])

# Add data labels
for i in range(len(players)):
    plt.text(run_5yrs[i] + 10, y[i], str(run_5yrs[i]), va='center', fontsize=9)

# Labeling
plt.xlabel("Runs")
plt.ylabel("Players")
plt.title("Total Runs Scored (2008–2012)")
plt.yticks(y, players)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
