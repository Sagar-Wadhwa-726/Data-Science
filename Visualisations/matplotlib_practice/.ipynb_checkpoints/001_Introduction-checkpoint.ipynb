```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000,2003,2005,2007,2010]
runs = [500,700,1100,1500,1800,1200,1700,1300,900,1500]
plt.plot(years, runs)
```

```python
# Multiple plots in a single graph
# Format string in matplotlub [color][line][marker]
kohli = [0,0,500,800,1100,1300,1500,1800,1900,2100]
sehwag = [0,300,800,1200,1500,1700,1600,1400,1000,0]
plt.plot(years, kohli, 'yo--', label = "Virat Kohli")
plt.plot(years, sehwag, 'g^:', label = "Virendar Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.title("Performance Comparision")
plt.show()
```

```python
# Making format strings more readable
kohli = [0,0,500,800,1100,1300,1500,1800,1900,2100]
sehwag = [0,300,800,1200,1500,1700,1600,1400,1000,0]
plt.plot(years, kohli, color='orange', linestyle = '--', label = "Virat Kohli")
plt.plot(years, sehwag, color = 'red', linestyle = '-.', label = "Virendar Sehwag")
plt.plot(years, runs, color='blue', label = "Sachin Tendulkar", linewidth = 0.4)
plt.tight_layout()
plt.grid(True)
plt.style.use('seaborn-v0_8-dark')

plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.title("Performance Comparision")
plt.show()
```

```python
# XKCD Comic style
with plt.xkcd():
    plt.plot(years, kohli, label = "Virat Kohli")
    plt.plot(years, sehwag, label = "Sehwag")
    plt.title("Epic Battle of the Batsmen")
    plt.legend()
    plt.show()
```
