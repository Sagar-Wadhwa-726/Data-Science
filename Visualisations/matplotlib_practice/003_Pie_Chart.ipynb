```python
import pandas as numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pie chart should be used only when we have lesser number of categories, other wise it will get cluttered
# If there are a lot of categories then consider some other form of chart
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F']
plt.pie(runs, labels=labels, colors=colors,
       wedgeprops = {
           'edgecolor' : 'blue',
           'linewidth' : '1.5',
           'linestyle' : '--',
           'fill' : True
       })
plt.title("Career runs of Indian Batsmen")
plt.show()
```

```python
# Explode will make a particular slice come out according to the value
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F']
explode = [0.1,0.5,0,0]

plt.pie(runs, labels=labels, colors=colors, explode=explode)
plt.title("Career runs of Indian Batsmen")
plt.show()
```
