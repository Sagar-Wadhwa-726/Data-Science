```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3,2,2,5] # hours in a day
colors = ['blue', 'red', 'yellow', 'pink']
plt.figure(figsize=(6,6))
plt.axis('equal')
plt.title('After School Activites')
plt.pie(time_spent, labels=activities, colors=colors, startangle=90, autopct='%1.1f%%')
```

```python
# The above pie chart can't analyze the time spent by a child by let's say watching TV over a week (over a particular time 
# period). This tells how does individual parts contribute over a time. As the time progresses, how is the time spent
# on different activities changes. Stack plot can be used to track all of this
```

```python
days = [1,2,3,4,5,6,7]
studying = [3,4,1,1,0.2,1,0]
playing = [2,2,1,1,2,3,2]
watching_tv = [2,1,2,2,1,1,1]
sleeping = [5,5,6,5,5,3,1]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize = (10,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels = labels, colors=colors, alpha = 0.8)
plt.legend(loc = "upper left")
plt.title("Students Time Distribution")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
```
