```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
```

```python
type(tips)
```

```python
tips.head(10)
```

```python
sns.lineplot(x="total_bill",y="tip", data=tips)
plt.title("Line Plot")
```

```python
sns.scatterplot(x="total_bill",y="tip", data=tips)
```

```python
sns.barplot(x="total_bill",y="tip", data=tips)
plt.title("Bar Plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
sns.boxplot(x="total_bill",y="tip", data=tips)
plt.title("Bar Plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
sns.boxenplot(x="total_bill",y="tip", data=tips)
plt.title("Bar Plot")
plt.xlabel("x")
plt.ylabel("y")
```
