```python
colors = ['red' if score < 50 else 'green' for score in exam_scores]

# The more the score the more big the circle
sizes = [score * 2 for score in exam_scores]

plt.scatter(study_hours, exam_scores, c=colors, s=sizes)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study v/s Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
colors = ['red' if score < 50 else 'green' for score in exam_scores]

# The more the score the more big the circle
sizes = [score * 2 for score in exam_scores]

# We can give another list also instead of c=exam_scores, to color the points on the basis of that new list

plt.scatter(study_hours, exam_scores, c=exam_scores, cmap='spring')

#Add labels
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}', (study_hours[i]+0.2, exam_scores[i]))

plt.colorbar(label="Score")
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study v/s Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# MULTIPLE SCATTER PLOTS IN A SINGLE GRAPH
class_a_hours = [2,4,6,8]
class_a_scores = [45,55,65,85]

class_b_hours = [1,3,5,7,9]
class_b_scores = [40,50,60,70,90]

plt.scatter(class_a_hours, class_a_scores, label="A", color="blue", marker='o', s=50)
plt.scatter(class_b_hours, class_b_scores, label="B", color="green", marker='s', s=50)
plt.legend()
plt.style.use('ggplot')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hours vs Exam Scores for Class A and B")
plt.tight_layout()

for i in range(len(class_a_hours)):
    plt.annotate(f'A-S{i+1}', (class_a_hours[i]+0.2, class_a_scores[i]))

for i in range(len(class_a_hours)):
    plt.annotate(f'B-S{i+1}', (class_b_hours[i]+0.2, class_b_scores[i]))    
plt.grid(True)
plt.legend()
```
