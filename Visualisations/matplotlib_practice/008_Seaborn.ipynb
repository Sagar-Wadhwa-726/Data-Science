```python
# Seaborn should be used when we don't want to minutely configure our graphs
# If we want to control/configure each and every small details in our graphs then use matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='whitegrid')
```

```python
x = np.array(list(range(1,20)))
y = np.sin(x)
sns.lineplot(x=x,y=y)
```

```python
# Step 1: Generate data
x = np.linspace(0, 2 * np.pi, 1000)  # 1000 points between 0 and 2π
y = np.sin(x)

# Step 2: Create DataFrame (seaborn works best with DataFrames)
df = pd.DataFrame({'x': x, 'sin(x)': y})

# Step 3: Plot
sns.set(style="whitegrid")
sns.lineplot(data=df, x='x', y='sin(x)', color='blue')

# Show the plot
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.show()
```
